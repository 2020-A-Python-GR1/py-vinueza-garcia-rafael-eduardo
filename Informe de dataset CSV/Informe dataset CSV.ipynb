## Descripcion del dataset
<p style="text-align: justify;">El dataset escogido contiene información acerca de los videos trending diarios que han existido a lo largo de varios meses en la plataforma de Youtube durante los años 2017 y 2018, fue tomado de la página de datasets Kaggle y se puede acceder a él mediante el enlace https://www.kaggle.com/datasnaek/youtube-new donde se encuentra disponible para su descarga. Este dataset contiene las columnas: video_id, trending_date, title,channel_title, category_id, publish_time, tags, views, likes, dislikes, comment_count, thumbnail_link, comments_disabled, ratings_disabled, video_error_or_removed y description. Existen otros archivos similares con la información de videos trending en Mexico, Japón, Korea, Rusia entre otros pero para este trabajo en particular de han escogido los datasets de Estados Unidos, Francia, Alemania y Gran Bretaña ya que no contienen caracteres extraños que imposibilitan la carga de datos como ocurrio el caso con el archivo de Mexico o Korea.</p>

## Carga de datos

<p style="text-align: justify;">Antes de empezar a trabajar con la informacion debemos cargar las librerias necesarias para manipular y mostrar la informacion ademas de cargar los datos de los archivos csv mediante las siguientes lineas de codigo:</p>

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

dataframe_videos_US = pd.read_csv('archivosCSV/USvideos.csv')
dataframe_videos_FR = pd.read_csv('archivosCSV/FRvideos.csv')
dataframe_videos_DE = pd.read_csv('archivosCSV/DEvideos.csv')
dataframe_videos_GB = pd.read_csv('archivosCSV/GBvideos.csv')
```

## Preguntas

### 1.- ¿Cuales son los 10 videos trending con mas visitas en Estados Unidos?

```python
dataframe_resultado = dataframe_videos_US.sort_values('views',ascending=False)
dataframe_resultado = dataframe_resultado.drop_duplicates(['title'], keep='last')
dataframe_resultado = dataframe_resultado.head(10)

titulos = dataframe_resultado['title'].map(lambda x: x.replace(' ', '')).unique()
titulos_grafico = np.array(['v_1','v_2','v_3','v_4','v_5','v_6','v_7','v_8','v_9','v_10'])
visitas = dataframe_resultado['views'].map(lambda x: x).unique()

fig, ax = plt.subplots()
ax.bar(titulos_grafico, visitas)
ax.set_ylabel('Numero de visitas [decenas de millones]')
ax.set_xlabel('Videos trending')
ax.grid(True)

plt.show()

for i in range(len(titulos)):
    print('v_' + str((i + 1)) + ' = ' + titulos[i])
```
